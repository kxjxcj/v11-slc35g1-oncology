# 01 — Data download (D10.1 scGen + D11.1 MD)

Downloads the **3 single-cell h5ad** datasets for D10.1 (Dixit 2016, Norman 2019, Frangieh 2021) and the **27 AlphaFold PDB files** for D11.1 MD.

All data are 100 % public (NCBI GEO + AlphaFold DB + ENA).  Re-running this notebook is safe — files already present in `data/` are skipped.

Total download size: ~6.6 GB.  Expected runtime on a 50 Mbps link: ~25 min.

## 1.1 — Set up paths

```python
from pathlib import Path
DATA_DIR = Path('../data')
DATA_DIR.mkdir(parents=True, exist_ok=True)
print(f'data dir: {DATA_DIR.resolve()}')
```

## 1.2 — Single-cell h5ad (Dixit 2016, Norman 2019, Frangieh 2021)

```python
# The download script lives in code/src/download_data.py.  It mirrors the
# official GEO supplementary archives plus the Figshare Norman 2019 archive.
# Re-running this cell is idempotent.
import subprocess, sys
result = subprocess.run(
    [sys.executable, '../code/src/download_data.py',
     '--out-dir', str(DATA_DIR), '--skip-if-exists'],
    capture_output=True, text=True,
)
print('stdout:', result.stdout[-500:])
print('stderr:', result.stderr[-500:])
print('exit', result.returncode)
```

## 1.3 — AlphaFold PDB v6 (27 proteins)

```python
result = subprocess.run(
    [sys.executable, '../code/src/download_alphafold_v6.py',
     '--out-dir', str(DATA_DIR / 'pdb'), '--skip-if-exists'],
    capture_output=True, text=True,
)
print('exit', result.returncode)
print('stderr tail:', result.stderr[-300:])
```

## 1.4 — Verify downloads

```python
import os
expected_h5ad = ['dixit_2016_raw_processed.h5ad',
                 'dixit_2016_processed_processed.h5ad',
                 'norman_2019_processed_processed.h5ad']
missing = [f for f in expected_h5ad
           if not (DATA_DIR / f).exists()]
if missing:
    print('MISSING:', missing)
else:
    print('✓ all 3 h5ad present')

pdb_dir = DATA_DIR / 'pdb'
if pdb_dir.exists():
    n_pdb = len(list(pdb_dir.glob('*.pdb')))
    print(f'pdb count: {n_pdb}/27')
else:
    print('pdb dir missing — re-run 1.3')
```
