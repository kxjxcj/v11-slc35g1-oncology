# 10 — Reproduce the 81 main figures from v10 final

Renders the 81 PNG figures from the cached CSV/JSON results.

Each figure is saved to `figures/main/` and named after the figure number in the manuscript (e.g. `fig01.png`, `fig02.png`, …, `fig81.png`).

Expected runtime: ~5 min.

```python
import os, json
from pathlib import Path
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

FIG = Path('../figures/main')
FIG.mkdir(parents=True, exist_ok=True)
print(f'figures dir: {FIG.resolve()}')
```

## 10.1 — D4 pan-cancer box plot (Fig. 1)

```python
import pandas as pd
df = pd.read_csv('../results/D4_pancan/D4_pancan_results.csv')
fig, ax = plt.subplots(figsize=(12, 6))
df.groupby('project_id')['SLC35G1_median_log2TPM'].median().sort_values().plot.barh(ax=ax)
ax.set_xlabel('SLC35G1 log2(TPM+1) median')
ax.set_title('SLC35G1 pan-cancer distribution (n=663, 33 cancers)')
plt.tight_layout()
plt.savefig(FIG / 'fig01.png', dpi=120)
plt.close()
print('wrote fig01.png')
```

## 10.2 — D5 PPI centrality (Fig. 30)

```python
centrality = pd.read_csv('../results/D5_ppi/D5_centrality.csv')
fig, ax = plt.subplots(figsize=(8, 6))
centrality.nlargest(15, 'degree').plot.bar(x='gene', y='degree', ax=ax, legend=False)
ax.set_title('Top 15 hub genes by degree')
plt.tight_layout()
plt.savefig(FIG / 'fig30.png', dpi=120)
plt.close()
print('wrote fig30.png')
```

## 10.3 — D8 VKO waterfall (Fig. 60)

```python
depmap = pd.read_csv('../results/D8_vko/D8_depmap_results.csv')
fig, ax = plt.subplots(figsize=(10, 6))
essential = depmap[depmap['is_essential'] == 1]
top_genes = essential.groupby('gene')['chronos_score'].mean().sort_values().head(20)
top_genes.plot.bar(ax=ax, color='steelblue')
ax.axhline(0, color='red', linestyle='--', alpha=0.5)
ax.set_title('Top 20 essential genes by DepMap Chronos (mean across cell lines)')
ax.set_ylabel('Chronos score (mean)')
plt.tight_layout()
plt.savefig(FIG / 'fig60.png', dpi=120)
plt.close()
print('wrote fig60.png')
```

## 10.4 — Render remaining 78 figures from cached data

The remaining figures (fig02-fig29, fig31-fig59, fig61-fig81) are auto-generated from the cached `results/` CSV/JSON.  See `code/workflows/Snakefile` for the full rendering pipeline.

```python
import subprocess, sys
result = subprocess.run(
    [sys.executable, '../code/workflows/render_remaining_figures.py',
     '--out-dir', str(FIG)],
    capture_output=True, text=True,
)
print('exit', result.returncode)
print('stderr tail:', result.stderr[-300:])
n_fig = len(list(FIG.glob('*.png')))
print(f'figures rendered: {n_fig}/81')
```
