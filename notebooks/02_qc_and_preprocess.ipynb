# 02 — QC and preprocess (D10.1)

Runs the scanpy QC pipeline on the 3 h5ad datasets.  Generates:

- `data/qc_summary.json` — per-cohort QC metrics
- `results/D10_scGen/qc_summary.csv` — combined table

Expected runtime: ~30 min on 8 cores.

```python
import subprocess, sys
from pathlib import Path
DATA_DIR = Path('../data')
OUT = Path('../results/D10_scGen')
OUT.mkdir(parents=True, exist_ok=True)

result = subprocess.run(
    [sys.executable, '../code/src/qc_pipeline.py',
     '--data-dir', str(DATA_DIR),
     '--out-dir', str(OUT)],
    capture_output=True, text=True,
)
print('exit', result.returncode)
print(result.stdout[-500:])
print('stderr:', result.stderr[-300:])
```

## 2.1 — Inspect QC summary

```python
import json, pandas as pd
summary = json.loads(Path('../results/qc_summary.json').read_text())
print(json.dumps(summary, indent=2))
```
