# 04 — D5 protein-protein interaction network

Re-creates the STRING + HuRI + AlphaFold-Multimer integration reported in v10 final §9-§10.

Generates:

- Centrality table (CD44 degree 88, STAT1 betweenness 0.220)
- Cross-platform Jaccard matrix (HM-52 zero-overlap disclosure)
- Louvain modularity summary

Expected runtime: ~5 min.

```python
import pandas as pd, json
from pathlib import Path
RES = Path('../results/D5_ppi')
centrality = pd.read_csv(RES / 'D5_centrality.csv')
print(centrality.sort_values('degree', ascending=False).head(10).to_string(index=False))
```

## 4.1 — Cross-platform Jaccard (HM-52 zero-overlap disclosure)

```python
jaccard = pd.read_csv(RES / 'D5_jaccard.csv', index_col=0)
print(jaccard.round(3))
```

## 4.2 — Louvain modularity

```python
summary = json.loads((RES / 'D5_summary.json').read_text())
print(f'louvain modules: {summary["louvain_modules"]}')
print(f'louvain modularity: {summary["louvain_modularity"]:.4f}')
print(f'sub-network edges (physical): {summary["sub_physical_edges"]}')
print(f'sub-network nodes (physical): {summary["sub_physical_nodes"]}')
```

## 4.3 — GO enrichment (sialic acid metabolism FDR)

```python
go = pd.read_csv(RES / 'D5_go_enrichment.csv')
sialic = go[go.description.str.contains('sialic', case=False, na=False)]
print(sialic[['term_id', 'description', 'fdr']].head(5).to_string(index=False))
```
